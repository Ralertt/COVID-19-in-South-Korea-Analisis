```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#Загружаем файлы из датасета
case = pd.read_csv("Case.csv")
time = pd.read_csv("Time.csv")
gender = pd.read_csv("TimeGender.csv")
```

```python
#Строим график с общим числом тестов COVID-19
plt.figure(figsize=(10, 5))
plt.plot(time['date'], time['test'], label='Всего тестов')
plt.plot(time['date'], time['negative'], label='Отрицательные')
plt.plot(time['date'], time['confirmed'], label='Положительные')
plt.title('COVID-19 Количество тестов - Дата')
plt.xlabel('Дата ')
plt.ylabel('Количество тестов (млн)')
plt.legend()
plt.xticks(time['date'][::10], rotation = 55)
plt.show()
```

```python
#Строим график с количеством инфецированных COVID-19
plt.figure(figsize=(10, 5))
plt.plot(time['date'], time['confirmed'], label='Положительные')
plt.plot(time['date'], time['released'], label='Вылечилось')
plt.plot(time['date'], time['deceased'], label='Умерло')
plt.title('COVID-19 Количество инфецированных - Дата')
plt.xlabel('Дата ')
plt.ylabel('Количество тестов')
plt.legend()
plt.xticks(time['date'][::10], rotation = 55)
plt.show()
```
